```python
df_comp = pd.read_csv(r"C:\Users\phna007\OneDrive - Arapahoe County Government\Documents 1\ArcGIS\Projects\MyProject23\GC_tobacco_compliance.csv", encoding='latin1')
df_comp
```

```python
df_comp['X'] = df_comp['X'].round(6)
df_comp
```

```python
df['X'] = df['X'].round(6)
df
```

```python
df2['Count'] = df2.groupby('license_no')['license_no'].transform('size')
df2
```
